# `Введение в базы данных`

`База данных` - совокупность данных, которые хранятся по установленной `схеме`, взаимодействие с которыми выполняются в соответствии с `правилами` средств моделирования данных.

___

## <a> CAP-теорема</a> 

В любой реализации распределенных вычислений возможно обеспечить не более двух из трех следующих свойств:

- `согласованность` данных

    - данные не противоречат друг другу

- `доступность`:

    - корректно осуществляется ЗАПРОС $\rightarrow$ ОТКЛИК

- `устойчивость к разделению` :

    - каждая часть системы работает независимо друг от друга

`Первичный ключ` - поле, которое используется для обеспечения уникальности строк в таблице

`Внешний ключ` (вторичный) — это столбец (или группа столбцов), используемый для связи данных между таблицами.

## <a> Типы отношений в таблице</a> :

- Один к одному - для объекта возможно только `одно` значение признака, для признака возможен только `один` объект

- Один ко многим - для объекта возможно `несколько`  значений признака, для каждого из признаков возможен только `один` объект

- Многие ко многим - для объекта возможно `несколько`  значений признака, для каждого из признаков возможен `несколько` объектов

## <a> Нормализация БД</a>

`Нормализация БД` — это процесс организации данных для :

- упрощения процесс выборки

- обеспечения целостность данных

- улучшения масштабируемости

- предотвращения избыточности

- предотвращения несогласованнх зависимостей
