## <center> <a id=3 >Пайплайн команд</a>

- **STDout** `|` **STDin** : рез-тат выполнения левой команды помещается в правую в качестве аргумента

- **STDout** `>`  **файл** : рез-тат выполнения левой команды записывается в файл (`полностью переписывая` содержимое файла)

- **STDout** `>>`  **файл** : рез-тат выполнения левой команды записывается в файл (`дополняя` содержимое файла)

```python
less file.txt | grep in 
ls > file.txt
ls >> file.txt
```

## <center> <a id=4 >Права доступа</a>

- `chmod` : Изменение прав доступа

`Категория прав` : 

|Аргумент | Описание|
|:--:|--|
|u|	изменение прав для владельца|
|g|	изменение прав для группы|
|o|	изменение прав для всех остальных|
|a|	изменение прав для всех категорий (пользователь, группа и все остальные)|

`Характер изменения прав` : 

|Аргумент | Описание|
|:--:|--|
+|	добавить права
-|	отнять права
=|	установить права

`Вид доступа` : 

|Аргумент | Описание|
|:--:|--|
r|	права на чтение
w|	права на запись
x|	права на исполнение (для директории — чтение каталога)

`Числова нотация прав доступа` : 

<img src='data/img/chmod.png' width= 500></img>

```python
chmod u+x file.txt # Добавляет право на исполнение владельцу
chmod o-r file.txt # Отнимает прво чтения у остальных пользователей
chmod g=wx file.txt # Устанавливает право записывать в файл и исполнять для группы пользователей
```

## <center> <a id=5 >Файловая система</a>

- /`bin` : Каталог с бинарными и исполняемыми файлами `пользователя`
- /`sbin` : Каталог с бинарными и сполняемыми файлами `системы`
- /`etc` : Каталог с файлами `конфигураций`
- /`dev` : Каталог с файлами `устройств`
- /`proc` : Каталог с информацией о  `процессах`, использовании системных ресурсов
- /`var` : Каталог с `часто изменяемыми` системными файлами 
    - /`log` : Каталог с  `логами` системы
    - /`lib` : Каталог с  файлами `БД` 
    - /`loc` : Каталог с  файлами `блокировок` 
    - /`mail` : Каталог с  файлами `почты` 
    - /`spool` : Каталог с  файлами `печати` 
    - /`run` : Каталог с  `идентификационными номерами процессов `
    - /`tmp` : Каталог с `временными` файлами 
- /`usr` : Каталог с `программами` пользователей
    - /`bin` : Каталог с бинарными и исполняемыми файлами `пользователя`
    - /`sbin` : Каталог с бинарными и сполняемыми файлами `системы`
    - /`local` : Каталог с  `настройками` пользователей
- /`home` : Каталог с `личными файлами` пользователей
- /`lib` : Каталог с `системными библиотеками` 
- /`opt` : Каталог с `дополнительными программами`(драыверы, игрыб софт ...)
- /`mnt` : Каталог с `дополнительными файловыми системами`Linux
- /`media` : Каталог с `съемными носителями`
- /`srv` : Каталог с файлами `серверов`
- /`run` : Каталог с файлами `id процессов`
- /`sys` : Каталог с `информацией о системе`

### **<center> Nano</center>**

- `nano` : Запуск без файла

- `nano` _имя файла_ : Запуск с файлом

___
### **<center> Vim</center>**

- `vim` : Запуск без файла

- `vim` _имя файла_ : Запуск без файла

- `ESC + i` : Режим редактирования

- `ESC + :` : Режим командной строки

    - `set number` : Назначение номера строк
        - `: номер строки` : Премещение курсора на строку с номером
    - `:$` : Перемещение курсора на последнюю строку
    
    - `:/`_эл-нт поиска_: Поиск элемента в тексте
    - `:x!` : Выход с сохранением
    - `:q! `: Выход без сохранения

## <center> <a id=8 > Bash-скрипты </a>



- $\boxed{1}$ Написание скрипта в файле :

    - `#!/bin/bash` : Выбор интерпретатора (укзывается в первой строке)

        - `echo` _текст_ : Вывод на экран текста

- $\boxed{2}$ Назначение прав :

    - `chmod `_права_  _имя файла_ 

- $\boxed{3}$ Запуск файла :

    - `./`_название файла_ 

    - `bash` _название файла_ 

```python
nano script.sh

#!/bin/bash
var1=$(( 5 + 5 ))
var2=$(( $var1 * 2 ))
echo "Result $var2"

chmod a+x script.sh

bash script.sh
```
