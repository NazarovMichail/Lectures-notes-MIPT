## <a id='1'> $$f(x) = ax + b$$ </a>
- ### `Гладкая` : нет разрывов, изломов
- ### `Не ограниченная` (при a != 0) 
- ### `Монотонная` (при a != 0) : постоянно убывает или возрастает
- ### `Не периодическая`

```python
a = 2
b = 5
f = lambda x: a*x + b
x = np.arange(-10,10)
y = f(x)
```

```python
fig = plt.figure(figsize=(5,2.5), facecolor=(0.8,0.8,0.8))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, c='r')
ax.set_xlim((0,10))
ax.set_ylim((0,20))
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.grid()
ax.set_title('$f(x) = ax + b$')
```

## <a id='05'> $$f(x) = \frac {1} {x} $$ </a>
- ### `Гладкая` (при x != 0) : нет разрывов, изломов
- ### `Не ограниченная` 
- ### `Монотонная`  : постоянно убывает
- ### `Не периодическая`

```python
f = lambda x: 1 / (x)
x = np.arange(6e-10,10, 0.1)
x2 = np.arange(-10,-8e-10, 0.1)

y = f(x)
y2 = f(x2)
```

```python
fig = plt.figure(figsize=(5,2.5), facecolor=(0.8,0.8,0.8))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, c='r')
ax.plot(x2,y2, c='r')

ax.set_xlim((-10,10))
ax.set_ylim((-10,20))
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.axvline(0, c='black')
ax.grid()
ax.set_title('$f(x) = \\frac {1}{x}$')
```

# <a id='diff'>$$Производная$$</a>
## $f'(x) = \frac {f(x + \triangle x) - f(x)}{\triangle x}$ 
$\triangle x$ - бесконечно малое приращение аргумента
___

## `Свойства` производной
- #### $f'(x) > 0 $ : Функция возрастает в точке x
- #### $f'(x) < 0 $ : Функция убывает в точке x
- #### $f'(x) = 0 $ : Точка x - экстремум функции
___

## `Признаки недифференцируемости` функции:
- #### В `вершинах углов` функции
- #### В точках `разрыва` функции
- #### В точках с `вертикальными асимптотами`





___
### <a id='cal'>Вычисление производной</a>
___

```python
import sympy as sp

x = sp.Symbol('x')
f = x**3+x**2

dy_dx = sp.diff(f, x)
print(f"f'(x) = {dy_dx}")

x0 = 3
value = dy_dx.subs(x,x0)
print(f"f'({x0}) = {value}")
```

___
### <a id='min'>Минимум функции</a>
___

```python
from scipy.optimize import minimize_scalar

f = lambda x: x**2+x-10
X = np.arange(-10,10,0.1)
minimized = minimize_scalar(f)
print(minimized)
```

```python
plt.plot(X, f(X), label='$X^2 + X - 10$')
plt.yticks((-10.25,80))
plt.xticks((-10,-0.5,10))
plt.legend(loc='best')
plt.grid()
```
