## Импорт модулей:
- `imageio` для импорта изображения
- `numpy` для работы с данными изображения
- `matplotlib.pyplot` для вывода изображения

```python
import imageio as im
import numpy as np
import matplotlib.pyplot as plt

img = im.v2.imread('data/numpy_image.jpg')
plt.imshow(img)
```

## Изображение как `массив numpy`
___

Изображение представляет собой трехмерный массив `[ y, x, rgb ]`  

<img src='data/pixel.png'></img>

```python
print(img.shape)
print(img[759,503,:])
```

___
Обрезка изображения задается первыми 2-мя элементами массива

___
Изменение цвета задается 3-м элементом массива

```python
print(img[220,220,:])
plt.imshow(img[:,:,2])
```

 ___
### Преобразование `цвета` в серые тона


___
1. Для каждого пикселя получить массив `средних значений между тремя цветами`


```python
mean_img_rgb = img.mean(axis=2)
print(f'1) Массив средних значений размером: {mean_img_rgb.shape}: {mean_img_rgb[:2,:2]} ')
plt.imshow(mean_img_rgb)
```

___
2. Привести массив к исходному виду `[ y, x, [r, g, b] ]`

```python

img_shape = mean_img_rgb.shape
reshaped_mean_img = mean_img_rgb.reshape(img.shape[0], img.shape[1],1)
print(f'2) Массив средних значений размером: {reshaped_mean_img.shape}: {reshaped_mean_img[:2,:2,:]} ')
```

___
3. Задать `одно и то же` среднее значение для каждого цвета

```python
reshaped_mean_img = reshaped_mean_img.astype(int)
gray_img = np.repeat(reshaped_mean_img,3, axis=2)

plt.imshow(gray_img)
print(gray_img[:2, :2, :])
```
